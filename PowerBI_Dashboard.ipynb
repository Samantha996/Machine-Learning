<H3> Step 2: Cleaning up our data</H3>
<p> To properly clean the data, we will do the following:

1) Delete the first three rows of the data.

2) Make the next row the header.

3) Remove any values in the Category column which does not contain a "Total".

4) Remove the total column.

5) Unpivot the columns to change the rows.

6) Rename the columns into "Sales" and "Budget".

7) Change the column with the date values into a date type column.</p>
